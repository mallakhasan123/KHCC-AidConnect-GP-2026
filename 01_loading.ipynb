```python
# Updated file path pointing to Google Drive
file_path = '/content/drive/MyDrive/KHCC AidConnect GP 2026/KHCC_Dataset_v4_Accurate 2.xlsx'

# Load all 3 sheets
patients = pd.read_excel(file_path, sheet_name='Patient Profiles (250)')
donors   = pd.read_excel(file_path, sheet_name='Donor Profiles (250)')
matches  = pd.read_excel(file_path, sheet_name='Match History (Interactions)')

print("✅ Dataset loaded successfully")
print(f"Patients table : {patients.shape[0]} rows, {patients.shape[1]} columns")
print(f"Donors table   : {donors.shape[0]} rows, {donors.shape[1]} columns")
print(f"Matches table  : {matches.shape[0]} rows, {matches.shape[1]} columns")
```

```python
print("=== PATIENT COLUMNS ===")
for col in patients.columns:
    print(" -", col)

print("\n=== DONOR COLUMNS ===")
for col in donors.columns:
    print(" -", col)

print("\n=== MATCH COLUMNS ===")
for col in matches.columns:
    print(" -", col)
```

```python
print("=== MISSING VALUES — PATIENTS ===")
print(patients.isnull().sum())
print(f"\nTotal patient nulls: {patients.isnull().sum().sum()}")

print("\n=== MISSING VALUES — DONORS ===")
print(donors.isnull().sum())
print(f"\nTotal donor nulls: {donors.isnull().sum().sum()}")

print("\n=== MISSING VALUES — MATCHES ===")
print(matches.isnull().sum())
print(f"\nTotal match nulls: {matches.isnull().sum().sum()}")
```

```python
print("Duplicate rows in Patients:", patients.duplicated().sum())
print("Duplicate rows in Donors  :", donors.duplicated().sum())
print("Duplicate rows in Matches :", matches.duplicated().sum())
```
